```python
import pandas as pd
import numpy as np
```

```python

df1019_csv = "Data/JC-201910-citibike-tripdata.csv"
df1119_csv = "Data/JC-201911-citibike-tripdata.csv"
df1219_csv = "Data/JC-201912-citibike-tripdata.csv"
df0120_csv = "Data/JC-202001-citibike-tripdata.csv"
df0220_csv = "Data/JC-202002-citibike-tripdata.csv"
df0320_csv = "Data/JC-202003-citibike-tripdata.csv"
df0420_csv = "Data/JC-202004-citibike-tripdata.csv"
df0520_csv = "Data/JC-202005-citibike-tripdata.csv"
df0620_csv = "Data/JC-202006-citibike-tripdata.csv"
```

```python
df1019 = pd.read_csv(df1019_csv)
df1119 = pd.read_csv(df1119_csv)
df1219 = pd.read_csv(df1219_csv)
df0120 = pd.read_csv(df0120_csv)
df0220 = pd.read_csv(df0220_csv)
df0320 = pd.read_csv(df0320_csv)
df0420 = pd.read_csv(df0420_csv)
df0520 = pd.read_csv(df0520_csv)
df0620 = pd.read_csv(df0620_csv)
```

```python
df0620.head()
```

```python
citi_bike_data = pd.concat([df1019, df1119, df1219, df0120, df0220, df0320, df0420, df0520, df0620], ignore_index=True, sort=False)
citi_bike_data.tail()
```

```python
citi_bike_data['gender'] 
```

```python
citi_bike_data['Gender'] = citi_bike_data['gender'].replace([0,1,2], ['unkown', 'male', 'female'])
citi_bike_data = citi_bike_data.drop ('gender',1)
```
